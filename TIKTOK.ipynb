**Cette fonction Convertie les M en million et le K en Mille**

```python
def convert_values(x):
  if isinstance(x, str):
    if 'M' in x:
      x = x.replace('M', '')
      x = float(x) * 1000000
    elif 'K' in x:
      x = x.replace('K', '')
      x = float(x) * 1000
  return x

for col in ['nbr_subscribes', 'views', 'likes', 'comments', 'share']:
    data_fusion[col] = data_fusion[col].apply(convert_values)

data_fusion.to_csv('/content/drive/MyDrive/Data_Fussion_Tiktok/tiktok_Fusion_converted.csv', index=False) #Enregistrement du fichier
```

**Prétraitement pour la segmentation**



**Standardiser les fonctionnalités pour le clustering**

```python
# Fonctionnalités à utiliser pour la segmentation
features = tiktok_Fusion_with_new_column[['nbr_subscribes', 'views', 'likes', 'comments', 'share', 'engagement_rate']]
```

```python
# Standardisation des fonctionnalités
scaler = MaxAbsScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features
```

**Étape 5 : regroupement optimal à l'aide de la méthode du coude**

```python
# Utilisez KMeans pour déterminer le nombre optimal de clusters
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2, 10), metric='silhouette', timings=False)
```

```python
# Ajuster le visualiseur
visualizer.fit(scaled_features)
visualizer.show()
```

**Étape 6 : Effectuer un clustering K-Means**



```python
# Effectuer un clustering KMeans
optimal_k = 3  # Ajuster en fonction de la méthode du coude
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
tiktok_Fusion_with_new_column['cluster'] = kmeans.fit_predict(scaled_features)

# Associer les clusters aux catégories d'influenceurs
cluster_mapping = {2:'Macro',0:'Nano',1:'Micro'}  # Ajuster en fonction des résultats du clustering
tiktok_Fusion_with_new_column['influencer_type'] = tiktok_Fusion_with_new_column['cluster'].map(cluster_mapping)
```

```python
tiktok_Fusion_with_new_column['cluster'].value_counts()
```

**Étape 7 : Visualiser les résultats du clustering**

```python
# Nuage de points des clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=tiktok_Fusion_with_new_column['nbr_subscribes'], y=tiktok_Fusion_with_new_column['engagement_rate'], hue=tiktok_Fusion_with_new_column['influencer_type'], palette="viridis")
plt.title("Clustering of Influencers")
plt.xlabel("Number of Subscribers")
plt.ylabel("Engagement Rate")
plt.legend(title="Influencer Type")
plt.show()
```

```python
# prompt: essaie d'entraine DBSCAN, affiche le nuance de point de la distribution des données et les metriques

import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns


# Apply DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=3)  # Adjust eps and min_samples as needed
tiktok_Fusion_with_new_column['dbscan_cluster'] = dbscan.fit_predict(scaled_features)

# Evaluate DBSCAN (if applicable)
if len(set(tiktok_Fusion_with_new_column['dbscan_cluster'])) > 1:  # Check for more than one cluster
    silhouette_avg = silhouette_score(scaled_features, tiktok_Fusion_with_new_column['dbscan_cluster'])
    ch_score = calinski_harabasz_score(scaled_features, tiktok_Fusion_with_new_column['dbscan_cluster'])
    db_score = davies_bouldin_score(scaled_features, tiktok_Fusion_with_new_column['dbscan_cluster'])

    print(f"Silhouette Score: {silhouette_avg}")
    print(f"Calinski-Harabasz Index: {ch_score}")
    print(f"Davies-Bouldin Index: {db_score}")
else:
    print("DBSCAN resulted in only one cluster.  Re-evaluate parameters (eps, min_samples).")

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=tiktok_Fusion_with_new_column['nbr_subscribes'], y=tiktok_Fusion_with_new_column['engagement_rate'], hue=tiktok_Fusion_with_new_column['dbscan_cluster'], palette="viridis")
plt.title("DBSCAN Clustering of Influencers")
plt.xlabel("Number of Subscribers")
plt.ylabel("Engagement Rate")
plt.legend(title="DBSCAN Cluster")
plt.show()

# Analyze the distribution of points within each cluster
print(tiktok_Fusion_with_new_column['dbscan_cluster'].value_counts())

#Further analysis can be done on clusters based on number of data points in each cluster

#Example
# for cluster_num in set(tiktok_data['dbscan_cluster']):
#     if cluster_num != -1:  #Exclude outliers
#         cluster_data = tiktok_data[tiktok_data['dbscan_cluster'] == cluster_num]
#         print(f"Analysis of Cluster: {cluster_num}")
#         print(cluster_data.describe())
#         #Add more specific analysis of different clusters here

```

```python
# prompt: essaie d'entraine clustering hierarchique, affiche le nuance de point de la distribution des données et les metriques

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from scipy.cluster.hierarchy import dendrogram, linkage

# Assuming 'scaled_features' and 'tiktok_Fusion_with_new_column' are defined from previous code

# Hierarchical Clustering
n_clusters = 3  # You might need to adjust this based on your data and dendrogram
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
tiktok_Fusion_with_new_column['hierarchical_cluster'] = agg_clustering.fit_predict(scaled_features)

# Calculate metrics for hierarchical clustering
silhouette_avg = silhouette_score(scaled_features, agg_clustering.labels_)
ch_score = calinski_harabasz_score(scaled_features, agg_clustering.labels_)
db_score = davies_bouldin_score(scaled_features, agg_clustering.labels_)

print(f"Hierarchical Clustering Metrics:")
print(f"Silhouette Score: {silhouette_avg}")
print(f"Calinski-Harabasz Index: {ch_score}")
print(f"Davies-Bouldin Index: {db_score}")


# Visualize the dendrogram
linked = linkage(scaled_features, 'ward') # You can try different linkage methods like 'complete', 'average', 'single'
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()


# Visualize hierarchical clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=tiktok_Fusion_with_new_column['nbr_subscribes'],
                y=tiktok_Fusion_with_new_column['engagement_rate'],
                hue=tiktok_Fusion_with_new_column['hierarchical_cluster'],
                palette="viridis")
plt.title("Hierarchical Clustering of Influencers")
plt.xlabel("Number of Subscribers")
plt.ylabel("Engagement Rate")
plt.legend(title="Hierarchical Cluster")
plt.show()

# Analyze the distribution of points within each cluster for hierarchical clustering
print(tiktok_Fusion_with_new_column['hierarchical_cluster'].value_counts())
```
